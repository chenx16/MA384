```python
top_ten_seller = train.groupby("merchant_id")['Purchase Count'].sum().sort_values(ascending=False).head(10)
```

```python
train.gender.replace(0,"female",inplace=True)
train.gender.replace(1,"male",inplace=True)
train.gender.replace(2,"unknown", inplace = True)
```

```python
rows = 4
cols = 3
axes=[]
fig=plt.figure(figsize=[15,15])
a = 0;
for seller_id in top_ten_seller.index:
    axes.append( fig.add_subplot(rows, cols, a+1))
    subplot_title=("Top"+str(a+1)+"Seller_number_purchases_count_dist")
    axes[-1].set_title(subplot_title)
    highest_merchant_user_info = train[train.merchant_id == seller_id]
    highest_merchant_user_info.pivot_table(index=["age_range"], columns="gender", values="Purchase", aggfunc=np.sum).plot(kind='bar', ax=axes[-1])
#     f = highest_merchant_user_info.pivot_table(index=["age_range"], columns="gender", values="Purchase Count", aggfunc=np.mean)
#     plt.bar(f,10)
    a+=1
fig.tight_layout()    
plt.show()
```
