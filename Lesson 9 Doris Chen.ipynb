```python
import pandas as pd

profile = pd.DataFrame()
profile['name']  = ['Hue','Ibu','Rao']
profile['ID'] = ['6518','1290','0141']
profile['major'] = ['MA','CS','EE']
profile['year']   = ['Y1','Y3','Y2']
print('profile')
print(profile)
print()

tests = pd.DataFrame()
tests['name'] = ['Cox','Hue','Hue','Ibu']
tests['T1'] = [67,52,70,82]
tests['T2'] = [63,51,74,88]
tests['T3'] = [65,58,73,81]
tests.index = ['4915','6518','8711','1290']
tests.index.rename('ID',inplace=True)
print('tests')
print(tests)
print()

quizzes = pd.DataFrame()
quizzes['name'] = ['Cox','Cox','Hue','Hue']
quizzes['Q1'] = [60,53,55,73]
quizzes['Q2'] = [69,57,52,77]
quizzes['Q3'] = [65,52,59,74]
quizzes.index = ['4915','7621','6518','8711']
quizzes.index.rename('ID',inplace=True)
print('quizzes')
print(quizzes)
```

```python
profile.reset_index()
```

```python
tests.reset_index()
```
