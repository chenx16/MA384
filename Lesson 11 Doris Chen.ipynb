```python
import pandas as pd
col = ['id', 'gender', 'age', 'occupation', 'zip']
df_users = pd.read_csv('Data/Movies/users.dat', 
                       sep='::', 
                       header=None,
                       names=col)
df_users.head(3)
```

```python
#df_ratings column names: 

col = ['id', 'movie_id', 'rating', 'timestamp']
df_ratings = pd.read_csv('Data/Movies/ratings.dat', 
                       sep='::', 
                       header=None,
                       names=col)
```

```python
#df_ movie column names:

col = ['movie_id', 'title', 'genres']

df_movies = pd.read_csv('Data/Movies/movies.dat', 
                       sep='::', 
                       header=None,
                       names=col)
```

(a) Load the user, rating and movie data files and merge the data into a single dataframe.

```python
df = pd.merge(df_ratings,df_users)
df.head()
```

```python
df = pd.merge(df,df_movies)
df.head()
```

(b) Remove movies with fewer than 250 ratings.

```python
df1 = df.groupby('movie_id').filter(lambda x : len(x)>=250)
print(df.shape[0])
print(df1.shape[0])
```

(c) Determine the average movie ratings by gender.

```python
df_ave = df1.pivot_table(index="title", columns="gender",values='rating')
df_ave.head().round(1)
```

```python
pd.pivot_table(df1, values='rating', index=['title'],columns =['gender'], aggfunc = 'mean').head().round(1)
```

(d) List the 10 most popular movies with male viewers. List the 10 most popular movies with female
viewers.

```python
df_ave.sort_values(by='M', ascending=False).head(10)
```

```python
df_ave.sort_values(by='F', ascending=False).head(10)
```

(e) Which movies would likely be poor choices for a date? Which movies would likely be good choices
for a date?

```python
df_ave['diff'] = abs(df_ave['F']-df_ave['M'])
df_ave
df_ave.sort_values(by='diff', ascending=False).head(10)
```

```python
df_ave['diff'] = abs(df_ave['F']-df_ave['M'])
df_ave
df_ave.sort_values(by='diff', ascending=True).head(10)
```
