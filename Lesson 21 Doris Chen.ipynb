```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from scipy.stats import ortho_group
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('Data/Seeds/seeds_preprocessed.csv')
df.head(3)
```

```python
df.dtypes
```

```python
attributes = df[['asymmetry','grove']]
attributes.head(3)
```

```python
km = KMeans(n_clusters=3)
km.fit(attributes)
```

```python
labels = km.predict(attributes)
labels
```

```python
colors = np.array(['blue','red','green','cyan','magenta','yellow','black'])
marker_size = 100
marker_alpha = 0.5
df['clusters'] = colors[labels]
df.plot.scatter(x='asymmetry',y='grove',c=df.clusters.values,s=marker_size,alpha=marker_alpha)
```
