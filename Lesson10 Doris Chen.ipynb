```python
# %load "Data/Grades/grades_script2.py"
# grades3.py

import numpy as np
import pandas as pd

df = pd.DataFrame()
np.random.seed(0)
df['ID']     = np.random.randint(100,999,size=8)
df['name']   = ['Abe','Das','Fox','Han',
                'Ibu','Lee','May','Rao']
df['course'] = ['MA384','CSSE490','CSSE490','MA384',
               'CSSE490','MA384','MA384','CSSE490']
np.random.seed(0)
df['ave']    = np.random.randint(67,99,size=8)
df
```

```python
pd.cut?


    
>>> pd.cut(np.array([1, 7, 5, 4, 6, 3]),
...        3, labels=["bad", "medium", "good"])
[bad, good, medium, medium, good, bad]
Categories (3, object): [bad < medium < good]    
```

```python
cutpoints = np.array([0,59,64,69,74,79,84,89,100])
grades = pd.cut(df.ave,cutpoints,labels = ["F","D","D+","C","C+","B","B+","A"])
grades
```

```python
df['grades']=grades
df
```

```python
#pd.qcut?
bin_labels_3 = ['fair', 'average', 'excellent']
df['performance'] = pd.qcut(df['ave'],
                              q=[0, .25, .75, 1],
                              labels=bin_labels_3)
```
